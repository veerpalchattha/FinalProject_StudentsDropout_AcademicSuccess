## Dropout Rate by Gender

```python
# Create a binary dropout indicator
df['is_dropout'] = df['target'].apply(lambda x: 1 if x == 'Dropout' else 0)

# Calculate dropout rate by gender
dropout_by_gender = df.groupby('gender')['is_dropout'].mean().sort_values()

# Plot
plt.figure(figsize=(6,4))
dropout_by_gender.plot(kind='bar')
plt.ylabel('Dropout Rate')
plt.title('Dropout Rate by Gender')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df['gender'], df['target'])
chi2, p, _, _ = chi2_contingency(contingency)
```

```python
chi2
```

```python
p
```

There is a clear relationship between the Gender and DropoutRate.

## Age: Statistical Test

```python
df['is_dropout'].unique()
```

```python
df['target'].unique()
```

```python
from scipy.stats import mannwhitneyu

dropouts = df[df['is_dropout'] == 1]['age_at_enrollment']
graduates = df[df['is_dropout'] == 0]['age_at_enrollment']

mannwhitneyu(dropouts, graduates)
```

## Correlation Strength (Effect Size)

```python
df.groupby('is_dropout', observed=True)['age_at_enrollment'].mean().plot(kind='bar')
```

```python
# Groups: is_dopout=1, is_dropout=0
# H0: is_dropout_mean_age = not_dropout_mean_age
# H1: is_dropout_mean_age != not_dropout_mean_age 
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(
    dropouts,
    graduates,
    equal_var=False  # Welchâ€™s t-test
)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.5f}")
```

A two-sample Welchâ€™s t-test was conducted to compare the mean age at enrollment between dropout and non-dropout students. The results indicate a statistically significant difference between the two groups (p < 0.05).

## Dropout Rate by Course (Top 10 Courses)

```python
# Top 10 courses by number of students
top_courses = df['course'].value_counts().head(10).index

df_top_courses = df[df['course'].isin(top_courses)]

# Dropout rate by course
dropout_by_course = (
    df_top_courses
    .groupby('course')['is_dropout']
    .mean()
    .sort_values(ascending=False)
)

# Plot
plt.figure(figsize=(10,5))
dropout_by_course.plot(kind='bar')
plt.ylabel('Dropout Rate')
plt.title('Dropout Rate by Course (Top 10 Courses)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Hâ‚€ Dropout status is independent of course
# Dropout status is associated with course
```

```python
# Binary dropout variable
df['is_dropout'] = (df['target'] == 'Dropout').astype(int)

# Contingency table
contingency = pd.crosstab(df['course'], df['is_dropout'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency)

print(f"Chi2: {chi2:.2f}, p-value: {p:.5f}, dof: {dof}")
```

```python
#How strong is the relationship
n = contingency.values.sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency.shape) - 1)))
cramers_v
```

Although the Chi-square test indicated a statistically significant relationship between course and dropout status, the effect size measured by CramÃ©râ€™s V (V = 0.26) suggests a weak to moderate association. This indicates that while course choice influences dropout likelihood, it is not the sole determinant of student outcomes.
